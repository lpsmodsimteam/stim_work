## 1 -- Code Parameters

The Gross code is a **bivariate bicycle (BB) code** over
F_2[x, y] / (x^12 - 1, y^6 - 1). Two trinomials define the parity-check matrices:

- `A(x,y) = x^3 + y + y^2`
- `B(x,y) = y^3 + x + x^2`

These generate the 72x144 matrices H_X = [A | B] and H_Z = [B^T | A^T].

Qubit layout (Stim convention):

| Range   | Block        | Size |
|---------|--------------|------|
| 0-71    | Data L-block | 72   |
| 72-143  | Data R-block | 72   |
| 144-215 | X-ancilla    | 72   |
| 216-287 | Z-ancilla    | 72   |

Every data qubit has a lattice address `(i, j)` with `i in [0,12)` and `j in [0,6)`,
mapped to a flat index via `i*m + j`. The lattice wraps toroidally.


```python
print(f'l={lpu.L}, m={lpu.M}  =>  l*m = {lpu.N_C} checks of each type')
print(f'A monomials: {lpu.A_EXPS}')
print(f'B monomials: {lpu.B_EXPS}')
print(f'Data qubits : {lpu.N_DATA}  (L-block 0..{lpu.N_C-1},  R-block {lpu.N_C}..{lpu.N_DATA-1})')
print(f'X-ancilla   : {lpu.N_C}   (indices {2*lpu.N_C}..{3*lpu.N_C-1})')
print(f'Z-ancilla   : {lpu.N_C}   (indices {3*lpu.N_C}..{lpu.N_TOTAL-1})')
print(f'Total       : {lpu.N_TOTAL}')
```

## 2 -- Qubit Index Helpers

Every physical qubit lives at a `(block, i, j)` address.
Four helper functions map `(i, j)` to flat indices; `qubit_coords()` inverts them.

| Function           | Block     | Index formula    |
|--------------------|-----------|------------------|
| `data_L(i, j)`     | Data L    | `i*m + j`        |
| `data_R(i, j)`     | Data R    | `l*m + i*m + j`  |
| `x_check_anc(i,j)` | X-ancilla | `2*l*m + i*m+j`  |
| `z_check_anc(i,j)` | Z-ancilla | `3*l*m + i*m+j`  |


```python
print('data_L(0,0)      =', lpu.data_L(0, 0))
print('data_R(0,0)      =', lpu.data_R(0, 0))
print('x_check_anc(0,0) =', lpu.x_check_anc(0, 0))
print('z_check_anc(0,0) =', lpu.z_check_anc(0, 0))
print()
print('Round-trip (index -> qubit_coords):')
for idx in [0, 71, 72, 143, 144, 215, 216, 287]:
    block, i, j = lpu.qubit_coords(idx)
    print(f'  Q[{idx:3d}]  {block:<12s}  (i={i:2d}, j={j})')
```

## 3 -- Tanner Graph and Verification

The **Tanner graph** is bipartite:
- 144 **variable nodes** (data qubits)
- 144 **check nodes** (72 X-ancilla + 72 Z-ancilla)
- 864 **edges** -- each data qubit connects to 6 checks

The graph is **6-regular**: every data qubit is in 3 X-checks and 3 Z-checks;
every ancilla connects to exactly 6 data qubits.

`verify_code_structure()` checks 5 properties:
regularity, check weight, edge count, CSS condition H_X * H_Z^T = 0 (mod 2),
and CNOT-layer collision-freedom.


```python
g = lpu.build_tanner_graph()
lpu.verify_code_structure()
```

## 4 -- Checks: Weight-6 Stabilisers

Each **check** is a weight-6 Pauli stabiliser on 6 data qubits.
X-check at site `(i,j)` touches 3 L-block qubits (A monomials) and
3 R-block qubits (B monomials) -- from H_X = [A | B].

`qubits_in_check(s, "X"/"Z")` returns the 6 data qubit indices for check `s` (0..71).


```python
lpu.print_check_enumeration('X', max_checks=72)
print()
lpu.print_check_enumeration('Z', max_checks=72)
```

```python
# Change s to inspect any check index 0..71
s = 0
i_c, j_c = s // lpu.M, s % lpu.M
print(f'Check s={s}  site ({i_c}, {j_c})')
print(f'  X-ancilla : {lpu.x_check_anc(i_c, j_c)}')
print(f'  Z-ancilla : {lpu.z_check_anc(i_c, j_c)}')
print(f'  X-check data qubits: {lpu.qubits_in_check(s, "X")}')
print(f'  Z-check data qubits: {lpu.qubits_in_check(s, "Z")}')
```

## 5 -- Edges: Tanner Graph Connections

Each **edge** is a `(data_qubit, ancilla_qubit)` pair -- the two endpoints of a
CNOT gate in the syndrome circuit.

`edges_of_qubit(d)` returns all 6 edges at data qubit `d`:
3 to X-ancilla (from H_X) and 3 to Z-ancilla (from H_Z).


```python
for d in [0, 35, 72, 107]:
    block, i, j = lpu.qubit_coords(d)
    print(f'\nQ[{d}]  {block}  ({i},{j})')
    for data, anc in lpu.edges_of_qubit(d):
        ab, ai, aj = lpu.qubit_coords(anc)
        print(f'  edge ({data:3d}, {anc:3d})  ->  {ab}({ai},{aj})')
```

## 6 -- Syndrome Extraction CNOT Cycles

One syndrome extraction **cycle** = 6 CNOT **layers**, one per monomial in A and B.
All 72 CNOT pairs in each layer execute simultaneously.

- Layers 0-2: A monomials `(x^3,y^0)`, `(x^0,y^1)`, `(x^0,y^2)`
- Layers 3-5: B monomials `(x^0,y^3)`, `(x^1,y^0)`, `(x^2,y^0)`

For X-checks: ancilla is CNOT control, data qubit is target.
For Z-checks: data qubit is control, ancilla is target.

Total CNOTs per syndrome round: 6 x 72 X-pairs + 6 x 72 Z-pairs = 864.


```python
lpu.print_cycle_schedule()
```

```python
print('Data qubit participation per layer:')
schedule = lpu.cnot_cycle_schedule()
for layer in schedule:
    x_data = sorted({d for _, d in layer['x_pairs']})
    z_data = sorted({d for d, _ in layer['z_pairs']})
    lbl = layer['monomial']
    print(f"  Layer {layer['layer']}  ({lbl}):  X-data {x_data[:4]} ...  Z-data {z_data[:4]} ...")
```

## 7 -- Query API

Three key query functions:

- `checks_of_qubit(d)` -- X and Z check indices containing data qubit `d`
- `edges_of_qubit(d)` -- all 6 (data, anc) edges at `d`
- `data_qubits_in_cycle_layer(k)` -- which data qubits fire in CNOT layer `k`


```python
d = 0
ch = lpu.checks_of_qubit(d)
x_ancs = [lpu.x_check_anc(s // lpu.M, s % lpu.M) for s in ch['X']]
z_ancs = [lpu.z_check_anc(s // lpu.M, s % lpu.M) for s in ch['Z']]
print(f'Data qubit {d} -- checks:')
print(f'  X-checks: {ch["X"]}  (ancilla {x_ancs})')
print(f'  Z-checks: {ch["Z"]}  (ancilla {z_ancs})')
print()
print('Cycle layer membership for qubit 0:')
schedule = lpu.cnot_cycle_schedule()
for li in range(6):
    dset = lpu.data_qubits_in_cycle_layer(li)
    mono = schedule[li]['monomial']
    print(f'  Layer {li}  ({mono}):  present = {0 in dset}')
```

## 10 -- Tanner Graph Plot (Subset)

`plot_connectivity(subset=[...])` draws the induced Tanner subgraph with a spring layout.

Node colours:
- **Blue** -- Data L-block (0-71)
- **Orange** -- Data R-block (72-143)
- **Red** -- X-ancilla (144-215)
- **Green** -- Z-ancilla (216-287)
- **Gold** -- LPU qubits (once set via `set_lpu_component()`)

The full 288-node graph is a hairball; a local neighbourhood is more legible.


```python
# L-block rows 0+1 and R-block rows 0+1, plus all their ancilla
local_data = list(range(0, 12)) + list(range(72, 84))
anc_set = set()
for d in local_data:
    for _, a in lpu.edges_of_qubit(d):
        anc_set.add(a)
subset = local_data + sorted(anc_set)
print(f'Plotting {len(subset)} qubits ({len(local_data)} data + {len(anc_set)} ancilla)')
lpu.plot_connectivity(subset=subset)
```

## 11 -- Logical Operators and LPU Seeds

The Gross code encodes **12 logical qubits**. Each has a pair of weight-12 operators:
Z_bar_k (logical Z) and X_bar_k (logical X).

The LPU (Appendix A.3) measures the group generated by
`X_bar_1, X_bar_7, Z_bar_1, Z_bar_7` (0-based indexing: logical qubits 0 and 6).

The **support** of these operators seeds the LPU qubit sets:
- `V_l` is built around supp(Z_bar_1) = log_Z[0]
- `V_r` is built around supp(Z_bar_7) = log_Z[6]

`get_lpu_sets()` computes all logical operators (~10 s on first call).


```python
sets = lpu.get_lpu_sets()   # computes logical operators on first call
print(f'Z_bar_1 support  (V_l seed, {len(sets["V_l_seed"])} qubits): {sets["V_l_seed"]}')
print(f'Z_bar_7 support  (V_r seed, {len(sets["V_r_seed"])} qubits): {sets["V_r_seed"]}')
print(f'X_bar_1 support  ({len(sets["lx0_seed"])} qubits): {sets["lx0_seed"]}')
print(f'X_bar_7 support  ({len(sets["lx6_seed"])} qubits): {sets["lx6_seed"]}')
```

```python
print('Check neighbours of V_l seed:')
print(f'  X-ancilla: {sets["U_l_seed_xanc"]}')
print(f'  Z-ancilla: {sets["U_l_seed_zanc"]}')
print()
print('Check neighbours of V_r seed:')
print(f'  X-ancilla: {sets["U_r_seed_xanc"]}')
print(f'  Z-ancilla: {sets["U_r_seed_zanc"]}')
```

## Summary and Next Steps

### What `gross_lpu_analysis_bravyi.py` provides

| Function                       | Purpose                              |
|--------------------------------|--------------------------------------|
| `qubits_in_check(s, "X"/"Z")`  | 6 data qubits for check s            |
| `checks_of_qubit(d)`           | X and Z checks for data qubit d      |
| `edges_of_qubit(d)`            | All 6 Tanner graph edges at d        |

| `lpu_component(name)`          | Qubit list for V_l, E_l, U_l, etc.  |
| `set_lpu_component(name, idx)` | Set component from Appendix A.3      |
| `adjacency_dict()`             | Full qubit adjacency dict            |
| `print_connectivity_summary()` | Text connectivity table              |
| `plot_connectivity(subset)`    | Tanner graph plot                    |
| `plot_torus_layout(highlight)` | 12x6 torus grid                    |

1. Open arXiv:2506.03094, go to **Appendix A.3**
